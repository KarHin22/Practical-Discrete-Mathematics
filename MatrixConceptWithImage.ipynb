###Question 4

```python
from google.colab import files
from PIL import Image

# Upload the first image
uploaded1 = files.upload()
image1_path = list(uploaded1.keys())[0]
image1 = Image.open(image1_path)

# Upload the second image
uploaded2 = files.upload()
image2_path = list(uploaded2.keys())[0]
image2 = Image.open(image2_path)

# Get the dimensions of the images
width1, height1 = image1.size
width2, height2 = image2.size

# Calculate the total width for the combined image
total_width = width1 + width2

# Create a new blank image with the combined dimensions
combined_image = Image.new('RGB', (total_width, max(height1, height2)))

# Paste the first image onto the combined image
combined_image.paste(image1, (0, 0))

# Paste the second image next to the first image
combined_image.paste(image2, (width1, 0))

# Export the combined image as a JPG
combined_image.save("combined_image.jpg")

# Display the combined image
plt.imshow(combined_image)
plt.axis('off')
plt.title('Combined Image')
plt.show()
```

```python
from google.colab import files
from PIL import Image

# Convert the combined image to grayscale
gray_image = combined_image.convert("L")

# Export the grayscale combined image as a JPG
combined_image.save("gray_image.jpg")

# Display the converted image
plt.imshow(gray_image)
plt.axis('off')
plt.title('Gray Image')
plt.show()
```
